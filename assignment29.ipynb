14. how is memory managed in python?
=   In Python, memory management is performed automatically and
    is managed by the Python memory manager. The memory manager is 
    responsible for allocating memory to objects and data structures, 
    as well as freeing up memory when it is no longer needed.
    Python uses a garbage collector to automatically manage memory.
    The garbage collector keeps track of all objects that are in use 
    in the program, and it frees up memory used by objects that are no 
    longer referenced by the program.
    Overall, Python's memory management is designed to be efficient
    and transparent, so that developers can focus on writing code without
    having to worry about low-level memory management tasks.
